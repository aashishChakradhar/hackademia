```python
import matplotlib.pyplot as plt
import numpy as np
import random
# lets learn about plots
```

```python
x = [random.randrange(1,101) for i in range(20)]
y = [random.randrange(1,101) for i in range(20)]
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('random patter')
```

```python
x = np.arange(-100,100)

'''
y = x**2/4*a (where a = 1)
'''
Y = lambda x: x**2 / 4
y = [Y(i) for i in x]
plt.grid()
plt.plot(x,y)
```

<h1> Scatter Plot </h1>

```python
students = ['ram','hari','sita','gita']
maths = [50,70,64,88]
science = [85,67,81,74]
plt.grid()
for marker in students:
    plt.scatter(maths,science,marker='x',color = 'red')

plt.xlabel('maths')
plt.ylabel('science')
```

```python
'''
task 1:
    wap that asks for N number of students to enter record
    first it asks the name of student, marks in mathematics, marks in science
    create dataframe at the end and then dump it as data_students.csv
task 2:
    then in another cell read the csv file from data_students.csv
    scatter plot the marks of each students in mathematics and science

'''
```

```python
import pandas as pd
N = int(input('Enter the number of student for data entry:'))
name = []
maths = []
science = []
for i in range(N):
    name.append(input("enter name:\t"))
    maths.append(int(input('enter score in maths:\t')))
    science.append(int(input('enter score in science:\t')))
variables = {
    'name':name,
    'maths':maths,
    'science':science,
}


data_frame = pd.DataFrame(variables)
data_frame.to_csv('data.csv',index = False)
```

<h2>Comparision of 2 graphs in matplot</h2>

```python
def stock_price_calc(x,m=4,c=10):
    return m*x+c

X=[i for i in range(1,13)]
Y_1=[stock_price_calc(i) for i in X]
Y_2=[stock_price_calc(i,m=-3,c=5) for i in X]
Y_3=[stock_price_calc(i,m=6,c=1) for i in X]
plt.plot(X,Y_1)
plt.plot(X,Y_2)
plt.plot(X,Y_3)

plt.legend(['2005','2025','2023'])
plt.xlabel('Months')
plt.ylabel('Stock Price')
```
